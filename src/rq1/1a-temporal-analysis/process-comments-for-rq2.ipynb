# setup

```python
import pandas as pd

# Load the CSV file
file_path = './rq1_comments_github_unclean.csv'
cleaned_file_path = './rq1_comments_github_clean.csv'
df = pd.read_csv(file_path)
```

# cleanup invalid comments

```python
# Remove rows with useless comments
useless_comments = ["[dead]", "[flagged]", ""]
df = df[~df["comment_text"].astype(str).str.strip().isin(useless_comments)]

# Remove empty comments
df = df[df['comment_text'].notna()]

# Drop duplicates based on both 'comment_id' and 'comment_text'
df = df.drop_duplicates(subset=['comment_id', 'comment_text'])
```

```python
# Export cleaned data
df.to_csv(cleaned_file_path, index=False)
print(f"Cleaned data saved to {cleaned_file_path}")
```

# filter only for comments after chatgpt release date

```python
# Step 2: Filter comments after ChatGPT release date (Nov 30, 2022)
df["comment_date"] = pd.to_datetime(df["comment_date"], errors='coerce')
cutoff_date = pd.Timestamp("2022-11-30")
df_filtered = df[df["comment_date"] > cutoff_date]
```

```python
# Export filtered data
filtered_file_path = '/content/drive/MyDrive/datasets/muict-naist-senior/rq1/rq1_comments_all_cleaned_after_chatgpt.csv'
df_filtered.to_csv(filtered_file_path, index=False)
print(f"Filtered data saved to {filtered_file_path}")
```
